```python
Q1 = dataset['median_house_value'].quantile(0.25)
Q3=dataset['median_house_value'].quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
upper_limit=Q3+1.5*IQR
lower_limit=Q1-1.5*IQR
print(upper_limit)
print(lower_limit)
```

```python
outliers=[x for x in dataset['median_house_value'] if x<lower_limit or x>upper_limit]
print(outliers)
len(outliers)
```

```python
removed_outliers = dataset[(dataset['median_house_value'] >= lower_limit) &
                           (dataset['median_house_value'] <= upper_limit)]

print(f"Min value after outlier removal: {removed_outliers['median_house_value'].min()}")
print(f"Max value after outlier removal: {removed_outliers['median_house_value'].max()}")
```

***ENCODING INDEPENDENT DATA***

***SPLITTING INTO TRAINING AND TESTING SET***

```python
X=dataset.drop(['median_house_value'],axis=1)
print(X)
```

```python
y=dataset['median_house_value']
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
y_test = np.array(y_test)
print(y_test)
```

***FEATURE SCALIING***


```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_test)
```

```python
print(X_train)
```

***TRAINING THE MULTIPLE LINEAR  REGRESSION MODEL ON TRAINING SET***

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

***PREDICTING THE RESULTS***

```python
y_pred = regressor.predict(X_test)
y_pred=np.array(y_pred)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
print('Training Score: ', regressor.score(X_train, y_train))
print('Testing Score: ', regressor.score(X_test, y_test))
```

```python
from sklearn.metrics import mean_squared_error, r2_score
R2 = r2_score(y_test, y_pred)
```
